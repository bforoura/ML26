# **Data Visualization**

* Data visualization is the bridge between raw numbers and human intuition in data science and machine learning.

* It allows you to instantly spot hidden patterns, identify anomalies like outliers or missing data, and diagnose how a machine learning model is performing far better than raw statistics ever could.

* Ultimately, it turns complex algorithmic outputs into clear, actionable stories, making it essential for both debugging your code and explaining your findings to non-technical stakeholders.

**1. import matplotlib.pyplot as plt**

 * **pyplot** is a specific module inside it that gives you a MATLAB-like interface for making graphs.

  * This line creates a shortcut or **alias**. Instead of typing matplotlib.pyplot.plot() every time you want a graph, you can just type plt.plot().


<br>

**2. import seaborn as sns**

  * **Seaborn** is built on top of **Matplotlib**.
  
  * It makes it much easier to create beautiful, statistically minded charts like heatmaps, plots, and scatter plots with a lot less code.


<br>

**3. import pandas as pd**
 * We covered this in the previous notebook.

 * Pandas is the ultimate tool for data manipulation and analysis.

 * It introduces the **DataFrame**, which we can think of as a super-powered SQL table or Excel spreadsheet right inside your Python code.


<br>

**4. from sklearn.datasets import load_iris**
 * **scikit-learn** (imported as sklearn) is Python's premier machine learning library.

 * Instead of importing the whole massive library, it grabs just one specific function: **load_iris**.

 * This function loads the famous Iris flower dataset—a classic dataset containing measurements (petal and sepal lengths) of 150 iris flowers across three species.

 * It is widely used by beginners to practice data visualization and classification algorithms.

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
```

```
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```
plt.hist(df['sepal length (cm)'])
plt.show()
```

```
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['target'])
plt.show()
```

```
sns.heatmap(df.corr(), annot=False)
plt.show()
```
