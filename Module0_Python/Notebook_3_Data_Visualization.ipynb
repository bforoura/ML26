```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
```

# **Drawing a Histogram**

* The following code cell creates a **histogram** of the "sepal length" column from our dataset.


  * **df['sepal length (cm)']**: This targets the specific column named '**sepal length (cm)**' inside the Pandas DataFrame (df).

  * **plt.hist**(...): This calls Matplotlib's histogram function. It takes all those sepal length numbers, groups them into ranges called "**bins**", and counts how many flowers fall into each range.

  * It then draws a bar chart showing the distribution of the data. The x-axis shows the measurements (e.g., 4cm to 8cm), and the y-axis shows the frequency (how often those lengths appear).
  
  * It instantly tells you if most flowers have short, medium, or long sepals.

```
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```
plt.hist(df['sepal length (cm)'])
plt.show()
```

# **Drawing a Scatter Plot**


* Let's now use **sns** to create a scatter plot to visualize the relationship between two different features of the iris flowers, while color-coding them by their species.


 * **sns.scatterplot(...)**: This calls the scatter plot function from the Seaborn library, which places individual data points on a 2D grid.

  * **x=df['sepal length (cm)']**: This puts the sepal length on the horizontal X-axis.

  * **y=df['sepal width (cm)']**: This puts the sepal width on the vertical Y-axis.

  * **hue=df['target']**: This is the magic ingredient. "Hue" tells Seaborn to color-code the dots based on another column—in this case, the target column (the flower species). Flowers of the same species will get the same color.


<br>
  
  **Why Use Seaborn when we have Matplotlib?**

  * We can absolutely can use Matplotlib for this! Seaborn is built directly on top of Matplotlib, so anything Seaborn can do, Matplotlib can technically do too.

 * However, we use **Seaborn** (sns) here because it makes complex plots much easier to write and better looking out of the box.



```
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['target'])
plt.show()
```

# **Correlation Heatmaps**

* A **Correlation Heatmap** is a visual grid used to see how strongly every column in our dataset relates to every other column.

image.png

<br>

**1. df.corr()**

 * Before drawing anything, this calculates the **Pearson correlation coefficient** between all numerical columns in our DataFrame.

 * It generates a grid of numbers between -1 and 1.

   * 1 means perfect positive correlation (as variable A goes up, variable B goes up).

   * -1 means perfect negative correlation (as variable A goes up, variable B goes down).

   * 0 means no relationship at all.



<br>


**2. sns.heatmap(..., annot=False)**


 * This takes that grid of correlation numbers and turns it into a color-coded picture.

  * Seaborn automatically assigns colors based on the number:
     * **bright colors for strong relationships**
     * **dark colors for weak ones**
  
  * This lets us spot patterns instantly without squinting at a wall of raw numbers.

  * **annot=False** tells Seaborn not to write the actual numeric correlation values inside the squares.
  
  * It keeps the heatmap clean and visual.
  
  * If you wanted to see the exact numbers (like 0.85 or -0.2), you would change this to **annot=True**.


<br>

**3. plt.show()**

 * Because Seaborn is built on top of Matplotlib, it relies on Matplotlib to actually render the window and display the plot on your screen.

 * This line tells Python: "Okay, the drawing is done, now show it to me."

```
sns.heatmap(df.corr(), annot=False)
plt.show()
```
