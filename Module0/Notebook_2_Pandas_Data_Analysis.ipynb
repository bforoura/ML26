# Pandas for Machine Learning

```
import pandas as pd
import numpy as np
```

```
# Load a sample dataset (Iris)
from sklearn.datasets import load_iris
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

## Basic Exploration

```
df.info()
df.describe()
```

## Filtering

```
df[df['sepal length (cm)'] > 5]
```
